```python
def rot_brute(s):
    print(f"SRC: {s}\n")
    print(f"{'SHFT':>4} | DATA")
    print("-" * 30)

    for i in range(1, 26):
        res = ""
        for c in s:
            if c.isupper():
                res += chr((ord(c) - 65 + i) % 26 + 65)
            elif c.islower():
                res += chr((ord(c) - 97 + i) % 26 + 97)
            else:
                res += c
        print(f"{i:>4} | {res}")
# 下面("")內測試你的密文吧!!!
rot_brute("")
```
